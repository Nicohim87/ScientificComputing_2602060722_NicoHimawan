```python
# Interpolation

from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [1, 3, 3, 5, 2, 2]

x_point = 1.5
```

```python
f = interp1d(x,y)
x_lin = np.linspace(0, 5, 100)
y_lin = f(x_lin)

y_ask = f(x_point)
print(y_ask)
```

```python
plt.plot(x,y,'ob')
plt.plot(x_lin, y_lin, 'b')
plt.plot(x_point, y_ask, 'ro')
plt.show()
```
